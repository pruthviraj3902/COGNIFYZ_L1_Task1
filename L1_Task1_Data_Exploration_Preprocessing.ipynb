## Step 1: Import Libraries & Load Dataset

- In this step, we import necessary libraries and load the dataset.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('Dataset.csv')
df.head()
```

## Step 2: Understanding the Dataset

- Here, we check the structure, column names, and basic information of the dataset.

```python
df.columns
df.info()
df.describe()
```

## Step 3: Checking Missing Values

- In this step, we identify missing values present in the dataset.

```python
df.isnull().sum()
```

## Step 4: Handling Missing Values

- We handle missing values to prepare the dataset for analysis.

```python
# Fill missing values (example)
df = df.fillna(method='ffill')

# Check again
df.isnull().sum()
```

## Step 5: Checking Data Types

- We verify data types of columns for consistency.

```python
df.dtypes
```

## Step 6: Dataset Shape

- This step shows the number of rows and columns in the dataset.

```python
print("Dataset Shape:", df.shape)
```

## Step 7: Aggregate Rating Distribution

- In this step, we analyze the distribution of the target variable "Aggregate rating" to identify any class imbalance.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.histplot(df['Aggregate rating'], bins=10)

plt.title("Distribution of Aggregate Rating")
plt.xlabel("Aggregate Rating")
plt.ylabel("Count")
plt.show()
```

```python
print(df['Aggregate rating'].value_counts())
```
